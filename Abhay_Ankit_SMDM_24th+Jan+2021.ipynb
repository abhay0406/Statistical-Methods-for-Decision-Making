```python
#3.1
# Ho == Moisture content = 0.35
# Ha == Moisture content>0.35
#  both the samples will be tested saperatly as both are independent observations
# alpha == 0.05 (as its not stated)
from scipy import stats
from scipy.stats import ttest_1samp, ttest_ind
```

```python
df =pd.read_csv(r'C:\Users\abhay\Downloads\A shingles.csv')
```

```python
df.head()
```

```python
# sample sixe for A == 36 and b== 31
ta_statistics, pa_value = ttest_1samp(df,0.35)
print ('one sample ttest for A \nta statistics: {0} pa_value: {1} '.format(ta_statistics,pa_value))
```

```python
AP = pa_value/2
print(AP)
```

```python
# P_value is more than and alpha and based on testing we are not able to reject the null hypothesis. We can conclude that the moisture
# level of sample A is equal to 0.35
```

```python
df1 =pd.read_csv(r'C:\Users\abhay\Downloads\B shingles.csv')
```

```python
df1.head()
```

```python
tb_statistics, pb_value = ttest_1samp(df1,0.35)
print ('one sample ttest for B \ntb statistics: {0} pb_value: {1} '.format(tb_statistics,pb_value))
```

```python
BP=pb_value/2
print(BP)
```

```python
# P_value is less than and alpha and based on testing we can to reject the null hypothesis. We can conclude that the moisture
# level of sample B is more to 0.35 with 95% confidence level.
```

```python
#3.2
#   Ho == Xbar of A = Xbar of B
#   Ha == Xbar of A is not equal to XBar of B
mydata =pd.read_csv(r'C:\Users\abhay\Downloads\A & B shingles-1.csv')
mydata.head()
```

```python
mydata.isnull().sum()
```

```python
tm_statistic, pm_value  = ttest_ind(mydata['A'],mydata['B'], equal_var = True, nan_policy = 'omit') 
print('tstats', tm_statistic)
print('p value', pm_value)
```

```python
#from the above test we can conclude that P value is more than alpha of 0.05 and we fail to reject Null hypothesis. so the mean
#of sample A is equal to mean of Sample B
```
